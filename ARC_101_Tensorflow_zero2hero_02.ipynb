```python
# The WRONG! way to change the image to a gray_scale image
gray_image = (data[:,:,0] + data[:,:,1] + data[:,:,2])//3
# Display the image, for gray scale images you have pass the cmap to 'gray'
plt.imshow(gray_image, cmap='gray')
plt.show()
```

**Why did this happen?**
> well because the type is `uint8`, when first added the values will overflow and start from 0 after it reaches the maximum of 255

```python
# Load the image
data = np.array(plt.imread('/kaggle/input/arc-tensorflow-zero2hero-data/images/sign.jpg'))


# Set the red channel in this part of the image to 0
data[0:1000, 0:1000, 0] = 0

# Set the green channel in this part of the image to 255
data[0:1000, 0:1000, 1] = 255

# Set the blue channel in this part of the image to 0
data[0:1000, 0:1000, 2] = 0

# Visualize the result
plt.imshow(data)
plt.show()
```

```python
# inspired by openCV library that says we must do the average but with a weight for each channel
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114]).astype(int)
```

```python
img = plt.imread('/kaggle/input/arc-tensorflow-zero2hero-data/images/t-shirt.jpg')
tshirt = rgb2gray(img)
```

```python
tshirt.shape
```

```python
plt.imshow(tshirt, cmap = plt.get_cmap('gray'))
plt.show()
```
